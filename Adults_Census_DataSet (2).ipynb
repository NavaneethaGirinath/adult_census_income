# About Data Set


> Title: **Adult Census Income Dataset**

>Purpose: The dataset is used to build predictive models to determine whether an individual's annual income exceeds $50,000 based on various demographic and socioeconomic attributes.

> Number of Instances: The dataset contains a total of 15 instances.

> Number of Features: There are 14 key features in the dataset.

> Target Variable: The target variable is the "income" column, which indicates whether an individual's income exceeds $50,000 or not.

> Data Types: The dataset includes a mix of categorical and numerical features, such as age, education, occupation, marital status, etc.

> Missing Values: The dataset may contain missing values that need to be handled during preprocessing.

> Potential Applications: The dataset can be used for binary classification tasks, data analysis, feature engineering, and understanding income-related patterns and demographics.


Key Features (Attributes):

*   age: Age of the individual.

*   workclass: Type of employment.

*   education: Highest level of education achieved.

*   education-num: Numeric representation of education.

*   marital-status: Marital status of the individual.

*   occupation: Occupation of the individual.

*   relationship: Relationship status.

*   relationship: Relationship status.

*   race: Race of the individual.

*   sex: Gender of the individual.

*   capital-gain: Capital gains for the individual.

*   capital-loss: Capital losses for the individual.

*   hours-per-week: Hours worked per week.

*   native-country: Native country of the individual.

*   fnlwgt: Weighted factor representing the number of people the census
     believes.

*   fnlwgt: Weighted factor representing the number of people the census
     believes the entry represents.

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=4)
x_sam,y_sam=smote.fit_resample(x_train,y_train)
```

```python
y_sam.value_counts()
```

```python
x_re_train,x_re_test,y_re_train,y_re_test=train_test_split(x_sam,y_sam,test_size=.30,random_state=42)
```

```python
scaler=StandardScaler()
scaler.fit(x_re_train)
x_re_train=scaler.fit_transform(x_re_train)
x_re_test=scaler.fit_transform(x_re_test)
```

```python
tr=[]
def build_model(model):
  model.fit(x_re_train,y_re_train)
  y_pred_model=model.predict(x_re_test)
  acc=accuracy_score(y_re_test,y_pred_model)
  print("Accuracy Score:",acc)
  print("Confusion matrix","/n",confusion_matrix(y_re_test,y_pred_model))
  print(classification_report(y_re_test,y_pred_model))
  tr.append(acc)
```
